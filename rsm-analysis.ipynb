```python
import numpy as np
import math
from scipy.stats import norm
```

```python
a = 15
b = 42
sigma = math.sqrt(1/29+1/a+1/b+1/28)
print(sigma)
OR = (29*b)/(28*a)
print(OR)
z = np.log(OR)/sigma
print(z)
p = 1-norm.cdf(z)
print(p)
```

```python
from scipy.stats import fisher_exact

# New table
table = [[29, 15],  # Drug group
         [28, 42]]  # Placebo group

# Two-sided Fisher's Exact Test
odds_ratio, p_value = fisher_exact(table, alternative='greater')

print(f"Odds Ratio = {odds_ratio}")
print(f"P-value = {p_value}")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the Excel file
df = pd.read_excel("/kaggle/input/rsm-isi/Dataset.xlsx", sheet_name="Sheet1")

# Separate the data
rsm_scores = df[df['RSM'] == 1]['Year 1 Score']
non_rsm_scores = df[df['RSM'] == 0]['Year 1 Score']

# Perform Welch’s t-test
t_stat, p_value = ttest_ind(rsm_scores, non_rsm_scores, equal_var=True)

print(p_value)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Optional: interpret significance
alpha = 0.05
if p_value < alpha:
    print("Result: Statistically significant difference")
else:
    print("Result: No statistically significant difference")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load data
df = pd.read_excel("/kaggle/input/rsm-isi/Dataset.xlsx", sheet_name="Sheet1")

# Filter only RSM students
rsm_df = df[df['RSM'] == 1]

# Group 1: > 2 years in RSM
rsm_gt_2 = rsm_df[rsm_df['No of Years'] >= 3.5]['Year 1 Score']

# Group 2: <= 2 years in RSM
rsm_le_2 = rsm_df[rsm_df['No of Years'] < 3.5]['Year 1 Score']

# Summary statistics
print("RSM >= 3.5 years:")
print(f"Mean: {rsm_gt_2.mean():.2f}")
print(f"Median: {rsm_gt_2.median():.2f}")
print(f"Count: {rsm_gt_2.count()}")
print()

print("RSM < 3.5 years:")
print(f"Mean: {rsm_le_2.mean():.2f}")
print(f"Median: {rsm_le_2.median():.2f}")
print(f"Count: {rsm_le_2.count()}")
print()

# Welch’s t-test
t_stat, p_value = ttest_ind(rsm_gt_2, rsm_le_2, equal_var=False)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.6f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Statistically significant difference")
else:
    print("Result: No statistically significant difference")
```
